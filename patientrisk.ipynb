
# Predicting Patient Health Risk Using a Neural Network (PyTorch)
In this project, we will make a model that looks at several factors such as patient age, heart rate, oxygen level, and temperature. The model will them predict if the patient is low risk (0) or high risk (1) through a NN. 

```python
import torch 
import torch.nn as nn
import torch.optim as optim

#we first create a wide set of patient data, which includes age, heart rate, oxygen level, and temperature. We will use this data to train our model to predict the risk score for each patient.
#[age, heart rate, oxygen level, temperature]
#NOTE: we add double brackets size our data is not linear but 2D (ie, we have number of patients as well as number of features)
x = torch.tensor([
    [25, 72, 98, 36.8],  # young, normal = low risk
    [40, 85, 96, 37.0],  # normal = low risk
    [70, 110, 90, 38.5], # older + bad vitals = high risk
    [65, 105, 91, 38.0], # high risk
    [30, 75, 99, 36.7],  # low risk
    [80, 120, 88, 39.0], # high risk
    [55, 95, 94, 37.8],  # borderline = high risk
    [22, 70, 99, 36.5],  # low risk
], dtype=torch.float32)

#next, we create the corresponding risk scores for each patient. We will use these risk scores to train our model to predict the risk score based on the patient data.
y = torch.tensor([
    [0],  # patient 1
    [0],  # patient 2
    [1],  # patient 3
    [1],  # patient 4
    [0],  # patient 5
    [1],  # patient 6
    [1],  # patient 7
    [0],  # patient 8
], dtype=torch.float32)

#We have a linear model which takes in the 4 features and outputs a single value (the risk score)
model = nn.Sequential(
    nn.Linear(4, 1),
    nn.Sigmoid()
)

#we must also account for loss, which is the difference between our predicted risk score and the actual risk score 
loss_fn = nn.BCELoss()
```

In order to train our model, we must follow the following steps:
1) Make a prediction using the model
2) Compare to the real answer
3) Compute our error 
4) Adjust the weights of the model to reduce the error
5) Repeat

```python
#we create an optimizer which will update the weights of our model based on the loss
optimizer = optim.Adam(model.parameters(), lr=0.01)

#we will train our model for 1000 iterations
for epoch in range(1000):
    predictions = model(x)
    loss = loss_fn(predictions, y)
    optimizer.zero_grad() #resets memory of previous updates 
    loss.backward() #calculates the gradients of the loss with respect to the model parameters
    optimizer.step() #updates the model parameters based on the gradients
    print(loss.item())
```

Observe that, as the model learns, the loss decreases over time, indicating that predictions are becoming more accurate. The overall trend moves downwards, showing that the model is improving its understanding of the data.

```python
print(model(x))

with torch.no_grad():
    predictions = model(x)

    predicted_classes = (predictions >= 0.5).float()

    print("Predicted probabilities:")
    print(predictions)

    print("Predicted classes:")
    print(predicted_classes)

    print("Actual labels:")
    print(y)
```

The model’s predicted probabilities are extremely close to either 0 or 1, indicating a high level of confidence in its classifications. When these probabilities are converted into predicted classes using a 0.5 threshold, they align almost perfectly with the actual labels, suggesting that the model has successfully learned the patterns in the training data. However, the extreme values (e.g., very close to 0 or 1) also indicate that the model may be overfitting, meaning it has effectively memorized the small dataset rather than learning generalizable relationships. This is expected given the limited number of training examples and highlights the importance of using larger and more diverse datasets for real-world medical applications.
